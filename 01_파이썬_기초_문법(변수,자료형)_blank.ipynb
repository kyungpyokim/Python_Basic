### (3) 예제

반복할 횟수 구하기: 100개의 게시물을 한 페이지에 30개씩 보여주려면 총 몇 페이지가 필요할까요?

````python
# 1. 변수 정의
item = 100
show_count = 30
# 2. 연산 설명
page = item // show_count
page += item % show_count // 10

print(page)
````

````python
## 수학 문제 
## 100 / 30 몫 3 나머지 10
## 30개씩 3페이지가 꽉 차고 다음 페이지에 10개의 게시물이 올라와야 한다. 
## 결론 4페이지를 만들어야 한다. 

# 1. 변수 정의
posts = 100
posts_per_page = 30

# 2. 연산 설명
pages = (posts // posts_per_page) + 1
remainder = posts % posts_per_page
print("100 나누기 30의 나머지", remainder)
print("나머지가 있기 때문에 몫에 1을 더해서", pages)

print("="*100)
## 만약에 120개의 게시물을 한 페이지에 30개씩 보여주려면 총 몇 페이지가 필요할까요
# 1. 변수 정의
posts = 120
posts_per_page = 30

# 2. 연산 설명
pages = (posts // posts_per_page)
remainder = posts % posts_per_page
print("120 나누기 30의 나머지", remainder)
print("나머지가 없기 때문에 그대로", pages)

## 자동화 한다. 
## 1. 조건 : 나머지가 0이니?
## 맞다 -> + 1 하지 않는다.
## 아니다 -> + 1을 해야 한다.
if(remainder != 0):
    pages += 1

print("pages", pages)
````

문제. 주민등번호 뒷자리에서 성별을 제외하고 *로 표시해주세요.

````python
# 1. 뒷자리를 추출한다.
# 2. 뒷자리의 길이를 구한다.
# 3. 주민등록번호 길이에서 뒷자리의 길이를 뺀다.
# 4. 뺀 만큼만 맨 앞에서부터 가져온다.
# 5. * 뒷자리의 길이만큼 반복한다.
# 6. 합친다.
regist_num = "901212-1325221"

## 1. 901212-1 을 추출한다.
## 2. *을 6번 반복한다.
## 1번과 2번을 합친다.
temp = regist_num[:8]
print(temp)
stars = "*" * 6
print(stars)
print(temp + stars)
````
