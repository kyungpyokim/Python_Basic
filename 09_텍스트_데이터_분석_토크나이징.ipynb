## 2) Kiwipiepy - Kiwi 형태소 분석기

```python
from kiwipiepy import Kiwi

kiwi = Kiwi()
result = kiwi.analyze("안녕하세요. 형태소 분석기 키위입니다.")
print(result)
# result = kiwi.tokenize("안녕하세요. 형태소 분석기 키위입니다.")
# print(result)
```

```python
# Point : 리스트 안에 어떤 요소가 몇 개 있는지 프린트하면서 파악할 수 있다.
print(result, type(result))
print(len(result))
print("="*50)
print(result[0])
print(type(result[0]))
print("="*50)
# 반복 변수 정의 : word
# 반복 출력 : result에서 하나씩 뽑아서 word에 담은 후 차례대로 출력
# 조건 

for token, score in result:
    print(f"token = {token}")
    for t in token:
        if t.tag[0] == "N":
            print(f"t = {t}")
            print(f"form = {t.form}, tag = {t.tag}")
            print("="*50)
```

```python
from kiwipiepy import Kiwi

kiwi = Kiwi()

text = "안녕하세요. 저는 형태소 분석기 입니다."

result = kiwi.analyze(text)
print(result)
print(len(result))
print("="*50)
# analyze 함수는 1개의 요소를 가지고 있다.
result_0 = result[0]
print(result_0)
print(len(result_0))
print("="*50)
# 튜플 안에는 요소가 2개 있다.
print(result_0[0]) # 내가 필요한게 여기 있다.
print(result_0[1])
print("="*50)
final_result = result_0[0]
print(f"final : {final_result}")
print(len(final_result))
```

```python
# 명사인 단어만 추출하기
## kiwi에서 명사는 tag가 N으로 시작한다.
## 조건 표현 방법 1 - tag[0] == "N" 표현 방법 2 - tag.startwith("N")
noun_list = []
for res in final_result:
    form = res.form
    tag = res.tag
    if(tag.startswith("N")):
        noun_list.append(form)
        print(res)
        print(f"form={form}, tag={tag}")

print("="*100)
print(noun_list)
```

## (4) 워드클라우드

### 정규표현식

```python
import re

text = "오늘!!!!!!!          ㄴㄴㄴㄴㄴㄴㄴ너무  ㅓㅑㅏㄴ 추워어어 very cold VERY COLD 010-0000-0000 ^^ ㅎㅇㅎㅇ"

# re.sub(패턴, 대체할 값, 대상): 대상에서 "패턴"을 파악한후 그 값을 "대체할 값"으로 바꾼다.
# 패턴: "[^0-9a-zA-Zㄱ-ㅎㅏ-가-힣\s]" : 숫자, 영어 소문자, 영어 대문자, 자음, 모음, 한글, 띄어쓰기가 아닌 것
new_text = re.sub("[^0-9a-zA-Zㄱ-ㅎㅏ-ㅣ가-힣\s]", "", text)
print(new_text)
```

```python
# 목표 : 데이터프레임을 단어 리스트로 만드는 것

# 0. 빈 리스트를 만든다. word_list, stop_words
# 1. 데이터프레임에서 text 값을 하나씩 뽑는다. -> 문장 sent
# 2. sent에서 필요없는 문자(특수문자, 이모지 등)을 없앤다. (패턴" [^0-9a-zA-Z가-힣\s]) clean_sent
# 3. 형태소 분석기로 문장을 단어 리스트로 뽑는다. -> result
# 4. word_list에 stop_words에 없는 result 요소들을 추가한다.
```

```python
from konlpy.tag import Okt

okt = Okt()
```

```python
text = "안녕하세요. 파이썬입니다."
result_morphs = okt.morphs(text)
print(result_morphs)
result_pos = okt.pos(text)
print(result_pos)
```

```python
# 0. 빈 리스트를 만든다. word_list, stop_words
word_list = []
stop_words = ["배민", "배달", "민족", "리뷰", "자주", "주문"]

# 1. 데이터프레임에서 text 값을 하나씩 뽑는다. -> 문장 sent
for sent in df_new["text"]:
    # 2. sent에서 필요없는 문자(특수문자, 이모지 등)을 없앤다. (패턴" [^0-9a-zA-Z가-힣\s])
    clean_sent = re.sub("[^0-9a-zA-Z가-힣\s]", "", sent)
    # 3. 형태소 분석기로 문장을 단어 리스트로 뽑는다.(조건: Noun) -> result
    ## 3-1. (단어, 품사) 쌍으로 결과를 출력한다.
    result = okt.pos(clean_sent)
    ## 3-2. 하나씩 뽑아서 품사가 Noun인지 확인한다.(반복문)
    sub_list = []
    for res in result:
        word = res[0]
        pos = res[1]
        # 3-3. word가 stop_words에 있으면 건너뛴다.
        if word in stop_words:
            continue

        ## 3-4. 품사가 Noun이면 sub list에 담는다.
        if pos == "Noun" and len(word) > 1:
            sub_list.append(word)

    print(sub_list)
    # 4. word_list에 stop_words에 없는 result 요소들을 추가한다.
    word_list.extend(sub_list)
    # print(f"{sent[:40]}")
    # print(f"{clean_sent[:40]}")
    # print(result)
    # print("="*100)
```

```python
print(word_list)
print(len(word_list))
```

### 단어 카운팅하기(Counter)

```python
from collections import Counter

example = ["님", "여", "여", "남", "남", "남", "응답없음"]
counter = Counter(example)
print(counter)
```

```python
# 기능이 많이 발생한 순서대로 n개 추출
count_most = counter.most_common(2)
print(count_most)
print(dict(count_most))
```
