# 1. 시리즈(Series)

````python
import pandas as pd

# 하루 방문자 수 데이터
visiters = pd.Series(
    data = [120, 150, 130, 170, 160],
    name = "방문자 수" # 옵션
)

print(visiters)
````

````python
print(visiters.index) # 인덱스 추출
print(visiters.values) # 값 추출
print(visiters.dtype) # 데이터 타입
print(visiters.name) # 시리즈 이름
print(visiters.size) # 데이터 갯수

````

## 3) 결측치

### (1) 결측치 발생

````python
import numpy as np
import pandas as pd

data = pd.Series([1, 2, np.nan, 5])
print(data)
````

예제
* 데이터가 10, 20, 30인 시리즈 생성, 인덱스는 부산, 울산, 대구으로 설정
* 데이터가 1, 2, 3, 4인 시리즈 생성, 인덱스는 광주, 부산, 울산, 대구으로 설정
* 데이터의 덧셈

````python
data_1 = pd.Series(
    data = [10, 20 ,30],
    index = ["부산", "울산", "광주"]
)
print(data_1)

data_2 = pd.Series(
    data = [1, 2, 3, 4],
    index = ["광주", "부산", "울산", "대구"]
)
print(data_2)
````

````python
# 매칭이 되지 않는 인덱스 데이터는 결측값이 된다.
print(data_1 + data_2)
````

### (2) 결측치 파악

````python
total = data_1 + data_2
print(total)
````

````python
# 데이터 정보 확인하기
total.info()
````

````python
# 각 줄마다 결측치인지 아닌지 판단하기
total.isna()
````

````python
# 결측치가 몇 개인가요?(bool - True(1), False(0))
print(total.isna().sum())
````

### (3) 결측치 채우기

````python
print(total)
````

````python
# 결측치를 0으로 채워주세요.
total.fillna(0)
````

````python
print(total)
````

### (4) 결측치 삭제

````python
print(total.dropna()) # total 덮어씌워지지 않는다.
````

````python
print(total)
````

## 4) 통계

````python
data = pd.Series(
    data=[17, 25, 33],
    index=["부산", "울산", "대구"]
)
print(data)
````

````python
print(data.sum())   # 열의 합
print(data.mean())  # 열 평균
print(data.min())   # 초소값
print(data.max())   # 최대값
print(data.std())   # 표준편차
print(data.var())   # 분산
````

````python
print(data.describe())
````

````python
import pandas as pd

df = pd.DataFrame(
    data=data_list,
    columns=["이름", "성별", "나이", "키"]
)
df
````
