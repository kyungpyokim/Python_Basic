## 3) 매개변수

```python
# 덧셈 계산기: 
## 입력: a, b 
## 실행: a + b 반환

def calc_add(a, b):
     return a + b

calc_add(1, 3)
```

```python
def calc(a, b, op = "+"):
    if op == "+":
        return a + b
    elif op == "-":
        return a - b
    elif op == "*":
        return a * b
    elif op == "/":
        if a == 0 or b == 0:
            return "0으로 나눌수 없습니다."
        return a / b
    else:
        return f"{op} 올바른 연산자를 사용해 주세요.(+ - * /)"
   
calc(1, 1, "=")
```

## 2) 파일 생성/읽기/추가

### 생성

```python
with open("./data/mytext.txt", mode="w", encoding="utf-8") as f:
    f.write("안녕")
```

### 읽기

```python
with open("./data/mytext.txt", mode="r", encoding="utf-8") as f:
    data = f.read()
    print(data)
```

### 추가

```python
with open("./data/mytext.txt", mode="a", encoding="utf-8") as f:
    f.write("반가워")
```

```python
aespa = ["카리나", "윈터", "지젤", "닝닝"]

# 윈터
# 카리나
# 지젤
# 닝닝

with open("./data/aespa.txt", mode="w", encoding="utf-8") as f:
    for member in aespa:
        f.write(member + "\n")
```
