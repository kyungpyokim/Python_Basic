## 3) 난수 배열 생성

```python
import numpy as np
np.random.seed(96)

# 0과 1사이의 실수 랜덤 배열 생성 np.random.rand(axis0, axis1, axis2)
random_arr1 = np.random.rand(2, 3)
print(random_arr1)

# 정규분포 (평균 0, 표준편차) np.random.randn(axis0, axis1, axis2)
# 평균이 0이고 표준편차가 1인 랜덤 배열 생성
random_arr2 = np.random.randn(2, 3)
print(random_arr2)

# 특정 범위 내 정수 랜던 배열 생성 np.random.randint(start, end, shape)
random_arr3 = np.random.randint(0, 10, (2, 3))
print(random_arr3)
```

# 3. 배열 속성

```python
# 배열 속성
import numpy as np

# 1차원 데이터
data_1d = np.array([10, 20, 30, 40])

# 2차원 데이터
data_2d = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

# 3차원 데이터
data_3d = np.array([
    [
        [255, 0, 0],
        [0, 255, 0]
    ],
    [
        [0, 0, 255],
        [255, 255, 0]
    ]
])
```

```python
my_tuple = (3,)
print(my_tuple, type(my_tuple))
```

```python
# 속성: 변수의 정보(표현: 변수.속성이름)
print(data_1d.shape) # 차원 형태
print(data_1d.ndim) # 차원 수
print(data_1d.size) # 모든 요소 개수
print(data_1d.dtype) # 데이터 요소들의 타입

print(data_2d.shape) # 차원 형태
print(data_2d.ndim) # 차원 수
print(data_2d.size) # 모든 요소 개수
print(data_2d.dtype) # 데이터 요소들의 타입

print(data_3d.shape) # 차원 형태
print(data_3d.ndim) # 차원 수
print(data_3d.size) # 모든 요소 개수
print(data_3d.dtype) # 데이터 요소들의 타입
```

```python
# 데이터 타입을 바꾸고 싶을 때(정수 -> 실수)
print(data_3d)
data_3d_float = data_3d.astype('float')
print(data_3d_float)
print(data_3d_float.dtype)
```

# 4. 배열 변환

```python
# shape를 변환
import numpy as np

data = np.arange(0, 15)
print(data)
print(data.size)

# shape를 변화 (3, 5), 배열.reshape()
data_reshape = data.reshape(3, 5)
print(data_reshape)

# 전치행렬 (행/열 전환) 배열.transpose()
data_transpose = data_reshape.transpose()
print(data_transpose)
data_transpose_2 = data_reshape.T
print(data_transpose_2)

# n차원 배열을 1차원으로 변환 배열.flatten()
data_flatten = data_transpose.flatten()
print(data_flatten)
```

# 5. 배열 연산

```python
# 연산: 같은 위치에 있는 요소끼리
# 연산할 때에는 shape가 같아야 한다.
data_1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
data_2 = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(data_1)
print(data_2)
```

```python
print("배열의 덧셈")
print(data_1 + data_2)
print('='*30)

print("배열의 뺄셈")
print(data_1 - data_2)
print('='*30)

print("배열의 곱셈")
print(data_1 * data_2)
print('='*30)

print("배열의 나눗셈")
print(data_1 / data_2)
print('='*30)

print("배열의 거듭제곱")
print(data_2 ** data_1)
print('='*30)
```

```python
# data_1 배열에 모든 요소에 10을 곱하고 싶다.
print(data_1 * 10)
```

```python
data_1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(data_1)
print(np.sum(data_1))
print(np.mean(data_1))
print(np.max(data_1))
print(np.min(data_1))
print(np.var(data_1))
print(np.std(data_1))
```

```python
# 2차원 배열 요약 통계
data1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print(data1)
print("axis=0 요약")
print(np.sum(data1, axis=0))
print("axis=1 요약")
print(np.sum(data1, axis=1))
```

```python
data2 = np.array([
    [
        [255, 0, 0],
        [0, 255, 0]
    ],
    [
        [0, 0, 255],
        [255, 255, 0]
    ]
])

print(data2)
print("="*30)

print("axis=0 요약")
print(np.sum(data2, axis=0))
print("="*30)

print("axis=1 요약")
print(np.sum(data2, axis=1))
print("="*30)

print("axis=2 요약")
print(np.sum(data2, axis=2))
```

```python
data_3 = np.array([
    [
        [255, 0, 0],
        [0, 0, 255]
    ],
    [
        [0, 0, 0],
        [255, 255, 255]
    ]
])

print(data_3)
print("="*30)

print("axis=0 요약")
print(np.sum(data_3, axis=0))
print("="*30)

print("axis=1 요약")
print(np.sum(data_3, axis=1))
print("="*30)

print("axis=2 요약")
print(np.sum(data_3, axis=2))
```
