# 1. 조건문

````python
# 나이를 입력하면 미성년자인지 아닌지 구분하기

# 1. 조건 설정: 나이가 19 미만인가? 
# 2. 변수 정의:
# 3. 설계: 19 미만이면 "미성년자입니다" 출력

age = 10

if(age < 19):
    print("미성년자입니다.")
else:
    print("성인입니다.")
    
````

````python
age = int(input("나이를 입력하세요."))
print(age)

if(age < 19):
    print("미성년자입니다.")
else:
    print("성인입니다.")
````

````python
# 합격 여부 판단하기
# 점수가 60점 이상이면 "합격" 그 외 "불합격" 출력

# 1. 변수
# 2. 조건
# 3. 설계

score = int(input("점수를 입력하세요."))

if score >= 60:
    print("합격")
else:
    print("불합격")
````

````python
# 등급에 따라 출력 (A-Perfect, B-Exellent, C-Good, 그 외는 '노력하세요.')

# 변수 정의
grade = 'A'

# 조건 : grade == "A", grade == "B", grade == "C", 그 외
if grade == 'A':
    print('Perfect')
elif grade == 'B':
    print('Excellent')
elif grade == 'C':
    print('Good')
else:
    print('노오력하세요.')
````

## 1) 연산자

문제. 나이가 10세 이상이고, 키가 120cm 이상일 때만 놀이기구 탑승

* 탑승 가능하면 "통과" 아니면 "탑승 불가"라고 출력하기

````python
age = 10
height = 120

if(age >= 10 and height >= 120):
    print("통과")
else:
    print("탑승 불가")

````

문제. 10만원 이상 구매하거나 "Gold" 등급이면 10% 할인

* 할인이 가능하면 "10% 할인 가능" 아니면 "10% 할인 불가" 출력

````python
# grade = "Silver"
grade = "Gold"
price = 100_000
# price = 90_000

print(grade, price)
if(price >= 100_000 or grade == "Gold"):
    print("10% 할인 가능")
else:
    print("10% 할인 불가")
````

문제 2. 학생의 성적에 따른 장학금 여부 결정

* GPA가 4.0 이상이면 100% 장학금을 받습니다.
* GPA가 3.5 이상이면 50% 장학금을 받습니다.
* GPA가 3.0 이상이면 30% 장학금을 받습니다.
* 그 외에는 장학금을 받을 수 없습니다.

````python
gpa = 3.0

if gpa >= 4.0:
    print("장학금 100%")
elif gpa >= 3.5:
    print("장학금 50%")
elif gpa >= 3.0:
    print("장학금 30%")
else:
    print("장학금 없음.")
````

문제 3. 특정 단어가 문장에서 없는지 확인

* search_word가 text에 포함되지 않으면 해당 단어가 문장에 없다는 메시지를 출력합니다.

````python
text = '''
    데이터 분석은 흥미롭고 중요한 분야입니다.
'''

search_word = '데이터'

if search_word in text:
    print("있다 있어")
else:
    print("없네 없어")
````
