### (2) break

````python
# while True:, count = 3 일때 멈추기
count = 0

# 조건: count == 3
while True:
    count += 1
    print(count)

    if count == 3:
        break
````

### (2) continue

````python
# while True:, count = 5일 때 멈추기, 짝수일 때는 출력하지 말기
# 변수 정의
count = 0

# 조건: count = 5, count % 2 == 0
while True:
    count += 1

    if count % 2:
        continue
    
    print(count)

    if count == 5:
        break
````

````python
# 예제: 1부터 10까지 더하기
## 계획세우기
## 변수 정의 : result = 0, num = 0
## num은 반복이 될 때마다 1씩 늘어난다.
## 반복이 될 때마다 result에 num을 계속 더하기

## 설계
num = 0
result = 0

while True:
    num += 1
    print(f"현재 number: {num}")
    print(f"result + number: {result} + {num}")
    result += num
    print(f"현재까지의 합: {result}")
    print("-" * 30)

    if num == 10:
        break
````

````python
## 변형해보기: 1부터 10까지 짝수만 더하기 / 홀수만 더하기

num = 0
result = 0

while True:
    num += 1
    print(f"현재 number: {num}")
    if num % 2 != 0:
        result += num
        print(f"result + number: {result} + {num}")
        print(f"현재까지의 합: {result}")
        
    print("-" * 30)

    if num == 10:
        break
````

````python
# 예제: 비밀번호가 맞을 때까지 입력하기
## 계획세우기
## 변수 정의 password = wanted
## 반복: 입력(input()), 조건: if password == "wanted" 반복 끝

## 설계
password = "wanted"

while True:
    input_text = input("비밀번호를 입력하세요.")

    if input_text == password:
        print("로그인 되었습니다.")
        break
````

````python
id_ = "wanted"
password = "wanted@!"
wrong_count = 0

while True:
    text = input("아이디/비밀번호를 입력하세요.")
    id, pw = text.split()

    if id_ == id and pw == password:
        print("로그인 되었습니다.")
        break
    else:
        print("비밀번호가 틀렸습니다.")
        wrong_count += 1

    if wrong_count == 3:
        break
````

## 4) List Comprehension

````python
num_list = [i for i in range(1, 11)]
print(num_list)
````

````python
num_list = [i for i in range(1, 11) if i % 2 == 0]
print(num_list)
````

````python
num_list = ["짝수" if i % 2 == 0 else "홀수" for i in range(1, 11)]
print(num_list)
````
