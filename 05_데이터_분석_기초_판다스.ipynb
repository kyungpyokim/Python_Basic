## 3) 결측치

### (1) 결측치 발생

````python
import numpy as np 
import pandas as pd 

data = pd.Series([1, 2, np.nan, 4])
print(data)
````

예제
* 데이터가 10, 20, 30인 시리즈 생성, 인덱스는 부산, 울산, 대구으로 설정
* 데이터가 1, 2, 3, 4인 시리즈 생성, 인덱스는 광주, 부산, 울산, 대구으로 설정
* 데이터의 덧셈

````python
data1 = pd.Series(
    data=[10, 20, 30],
    index=["부산", "울산", "대구"]
)
print(data1)

data2 = pd.Series(
    data=[1, 2, 3, 4],
    index=["광주", "부산", "울산", "대구"]
)
print(data2)
````

````python
# 매칭이 되지 않는 인덱스 데이터는 결측값이 된다.
print(data1 + data2)
````

### (2) 결측치 파악

````python
total = data1 + data2
print(total)
````

````python
# 데이터 정보 확인하기 
total.info()
````

````python
# 각 셀마다 결측치인지 아닌지 판단하기 
total.isna()
````

````python
# 결측치가 몇 개인가요?(bool - True(1), False(0))
print(total.isna().sum())
````

````python
# 성별이 M인 데이터 조회
# 성별 추출 : df["성별"]
df.loc[  df["성별"]=="M"  ,  : ]
````

````python
# 키가 170이 넘는 사람 조회
# 키 추출: df["키"]
# 조건: df["키"] > 170
df.loc[ df["키"] > 170 , :]
````

````python
## startswith() / endswith()
text = "송윤지"
# 조건: 맨 앞에 "송"이 있는가?
print(text.startswith("송"))
# 조겅: 맨 뒤에 "송"이 있는가?
print(text.endswith("송"))
````

````python
# 이름이 "송"씨인 사람 조회
# 이름 추출: df["이름"], 
# 조건: df["이름"].str.startswith("송")
# (Hint: startswith())
df.loc[ df["이름"].str.startswith("송") , :]
````

````python
# 이름에 "수"가 있는 사람만 조회 
# 이름 추출: df["이름"], 
# 조건: df["이름"].str.contains("송")
# (Hint: startswith())
df.loc[ df["이름"].str.contains("수") , :]
````

````python
# 20대인 사람만 조회
# 나이 추출: df["나이"]
# 조건: df["나이"] >= 20, df["나이"] <30
# 조건 결합: AND(&)
df.loc[  (df["나이"] >= 20) & (df["나이"] <30) , :]
````

````python
# 여자이거나 키가 170cm 이상인 사람만 조회
# 성별 추출, 키 추출 : df["성별"], df["키"]
# 조건: df["성별"] == "F", df["키"] >= 170
# 조건 결합: OR(|)
df.loc[ (df["성별"] == "F") | (df["키"] >= 170)  , :]
````

````python
data1 = pd.DataFrame(
    data=[["강남", 1, 2], ["서초", 4, 5], ["노원", 5, 6]],
    columns=["지역명", "지점수", "매출"]
)
data1
````

````python
data3 = pd.DataFrame(
    data=[["강남", 10, 20], ["도봉", 40, 50], ["노원", 50, 60]],
    columns=["지역명", "방문자수", "직원 수"]
)
data3
````

````python
# data1 + data2 
# 이 결과가 우리가 원하는 것일까?
data1 + data2
````

````python
# 문제: 똑같은 테이블이 2개가 작성되어 있다.
# 의도: 지역별로 겹치는 것만 지점수, 매출을 합산하고 싶다. 
# 인덱스를 지역명으로 바꿔야 합니다. 
data1_loc_idx = data1.set_index("지역명")
data1_loc_idx
````

````python
data2_loc_idx = data2.set_index("지역명")
data2_loc_idx
````

````python
data1_loc_idx + data2_loc_idx
````

````python
# 문제: 지역별로 새로운 열이 추가되었다. 
# 의도: 지역별로 지점수, 매출, 방문자수, 직원수를 합치고 싶다. 
````

````python
# pd.merge()
total_data = pd.merge(
    left=data1,
    right=data3,
    how="outer",  # left, right, inner, outer
    on="지역명" # 누구를 기준으로
)
total_data
````

````python
# how: 가져올 on(=지역명)
## left: data1의 지역명에 맞춰서
## right: data3의 지역명에 맞춰서
## inner: data1, data3에 모두 있는 지역명에 맞춰서
## outer: data1, data3에 있는 모든 지역명에 맞춰서
````

## Q5. total 연비의 평균보다 높은 자동차는 PASS, 낮은 자동차는 FAIL로 구분하세요.

````python
df.head()
````

````python
total_mean = df["total"].mean()
print(total_mean)
````

````python
# 새로운 열 result : total > total_mean -> True or False 가 저장된다. 
df["result"] = (df["total"] > total_mean)
df.head()
````

````python
# result가 True인 것만 추출
df.loc[ df["result"] == False , :]
````

````python
df["result"] = df["result"].replace({True: "pass", False: "fail"})
df.head()
````

````python
import numpy as np 

df["result"] = np.where(df["total"] > total_mean, "pass", "fail")
df.head()
````

````python
# apply() : 시리즈의 하나 하나 데이터마다 함수를 적용시킨다. 
# list comprehension : 한줄 코드
# df["total"] 데이터 요소 하나하나를 x라는 변수라고 하자. 
# 만약에 x > total_mean , pass, 아니면 fail
# new_total = []
# if x > total_mean:
#     new_total.append("pass")
# else:
#     new_total.append("fail")

# 한줄코드: ["pass" if x > total_mean else "fail" for x in df['total']]d
````

````python
df["result"] = df["total"].apply(lambda x: "pass" if x > total_mean else "fail")
df.head()
````

````python
# 새로운 열 : 조건이 있다. 
# 1) True/False 열을 만들고, replace 로 대체
# 2) np.where() 바로 만드는 방법
# 3) df.apply() 만드는 방법
````
