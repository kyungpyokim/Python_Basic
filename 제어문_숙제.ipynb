```python
# 편의점 영업: 9시부터 22시 사이면 "영업 중", 그 외에는 "영업 종료"를 출력하세요.
def is_working(time: int):
    if time >= 9 and time <= 22:
        print("영업중")
    else:
        print("영업종료")

is_working(10)
is_working(23)
```

```python
# 택배비 계산: 물건 가격이 5만원 이상이면 "무료 배송", 아니면 "3000원 추가"를 출력하세요.
def ship_price(price: int):
    if price >= 50_000:
        print("무료배송")
    else:
        print("3000원")

ship_price(50000)
ship_price(3000)
```

```python
# 지하철 요금: 나이가 6세 미만이면 "무료", 6~12세는 "600원", 13세 이상은 "1250원"을 출력하세요.
def get_charge(age: int):
    if age < 6:
        print("무료")
    elif age >= 6 and age <= 12:
        print("600원")
    elif age >= 13:
        print("1250원")

get_charge(5)
get_charge(10)
get_charge(13)
```

```python
# 학점 부여: 시험 점수에 따라 90점 이상 'A', 80점 이상 'B', 70점 이상 'C', 나머지는 'F'를 출력하세요.
def get_grade(score: int):
    if score >= 90:
        print("A")
    elif score >= 80:
        print("B")
    elif score >= 70:
        print("C")
    else:
        print("F")

get_grade(90)
get_grade(80)
get_grade(70)
get_grade(60)
```

```python
# 비밀번호 확인: 설정된 비밀번호와 입력한 비밀번호가 일치하면 "로그인", 아니면 "실패"를 출력하세요.
def is_same_password(pw: str):
    if(pw == '1234'):
        print("로그인")
    else:
        print("실패")

is_same_password("1234")
is_same_password("0000")



```

```python
# 놀이공원 야간개장: 나이가 15세 이상이거나, 동반자 여부가 참(True)이면 "입장 가능"을 출력하세요.
def is_enter_park(age: int, company: bool):
    if age >= 15 or company:
        print("입장 가능")
    else:
        print("입장 불가")

is_enter_park(13, True)
is_enter_park(15, False)
is_enter_park(12, False)
```

```python
# 장학금 선발: 성적이 3.5 이상이고 봉사시간이 20시간 이상일 때만 "장학금 대상"을 출력하세요.
def is_scholarship(score: float, service_time: int):
    if(score >= 3.5 and service_time >= 20):
        print("장학금 대상")
    else:
        print("장학금 대상 아님")

is_scholarship(3.5, 20)
is_scholarship(3, 20)
is_scholarship(3.5, 18)
```

```python
# 주말 판별: 요일 이름이 "토요일" 또는 "일요일"이면 "늦잠 가능", 아니면 "출근"을 출력하세요.
def go_to_bed(day_week: str):
    if day_week == "토요일" or day_week == "일요일":
        print("늦잠 가능")
    else:
        print("어여 출근")

go_to_bed("토요일")
go_to_bed("일요일")
go_to_bed("수요일")
```

```python
# 버스 할인: 나이가 19세 미만이거나 65세 이상인 경우에만 "할인 대상"을 출력하세요.
def bus_discount(age: int):
    if age < 19 or age >= 65:
        print("할인 대상")
    else:
        print("할인 대상 아님")

bus_discount(18)
bus_discount(65)
bus_discount(20)
```

```python
# 미세먼지 경보: 미세먼지 농도가 030이면 "좋음", 3180이면 "보통", 그 외에는 "나쁨"을 출력하세요.
def dust(density: int):
    if density == 30:
        print("좋음")
    elif density == 3180:
        print("보통")
    else:
        print("나쁨")

dust(30)
dust(3180)
dust(400)
```

```python
# 식재료 확인: 냉장고 리스트 안에 내가 찾고자 하는 재료가 있는지 확인하여 출력하세요.
def refigerator(food: str):
    frige = ["김치", "참치", "고등어"]
    if food in frige:
        print("있다 있어")
    else:
        print("없네 없어")

refigerator("김치")
refigerator("볶음김치")

```

```python
# 욕설 필터링: 특정 단어 리스트에 사용자가 입력한 단어가 포함되어 있으면 "차단된 단어"를 출력하세요.

text = "안녕하세요. 이것은 나쁜욕입니다."

if '나쁜욕' in text:
    print("차단된 단어")
```

```python
# 당첨 확인: 당첨 번호들이 담긴 리스트에 내 번호가 있는지 확인하여 "1등 당첨"을 출력하세요.
win_list = [10, 11, 29, 34, 50]
my_num = 29

if my_num in win_list:
    print("1등 당첨")
```

```python
# 회원 가입: 탈퇴한 회원 리스트에 현재 가입하려는 아이디가 없는(not in) 경우만 "가입 가능"을 출력하세요.
id_list = ["aaa", "bbb", "ccc"]
id = "ddd"
if id not in id_list:
    print("가입 가능")
```

```python
# 대출 승인: 신용등급이 3등급 이내이면서 연봉이 일정 금액 이상인 경우에만 "승인"을 출력하세요.
def avail_loan(grade: int, income: int):
    if grade <= 3 and income >= 5000:
        print("승인")
    else:
        print("비승인")

avail_loan(3, 5300)
avail_loan(3, 4000)
avail_loan(4, 5000)
```

```python
# 잔액 확인: 원하는 물건이 상점 리스트에 있고, 내 지갑의 돈이 물건 가격보다 많아야 "구매 완료"를 출력하세요.
stuff_list = [10_000, 24_000, 100_000]

if stuff_list[0] <= 20_000:
    print("구매 완료")
```
